# Звіт із продажів
### У цьому розділі ми рахуємо підсумки за простими данними

```python
sales = [120, 135, 100, 150, 90, 200, 170]
total = sum(sales)
average = total / len(sales)
maximum = max(sales)
minimum = min(sales)

print("Загальна кількість продажів:", total)
print("Середнє значення:", average)
print("Максимальне значення:", maximum)
print("Мінімальне значення:", minimum)
```
