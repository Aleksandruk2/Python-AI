# Завдання 3

```python
from collections import Counter

votes = ["apple", "banana", "apple", "orange", "banana", "apple", "apple", "apple", "orange"]

counter = Counter(votes)

leader = counter.most_common(1)

print("Фрукт\tКількість")

for fruit, count in counter.items():
    print(f"{fruit}\t{count}")

print("Фрукт-лідер:", leader[0][0])
print("Кількість:", leader[0][1])

top3 = counter.most_common(3)

print("Топ-3:")

for fruit, count in top3:
    print(f"{fruit}: {count}")
```

## Завдання 4

```python
math = [80, 92, 75, 60, 88]
physics = [70, 65, 90, 85, 78]
english = [95, 100, 85, 90, 87]

subjects = {
    "Math": math,
    "Physics": physics,
    "English": english
}

print("Предмет\t\tСередній\tМінімальний\tМаксимальний")

for subject, scores in subjects.items():
    average = sum(scores) / len(scores)
    minimum = min(scores)
    maximum = max(scores)

    print(f"{subject}\t\t{average:.2f}\t\t{minimum}\t\t{maximum}")


best_subject = max(subjects, key=lambda x: sum(subjects[x]) / len(subjects[x]))
print("\nПредмет з найвищим середнім балом:", best_subject)
```
